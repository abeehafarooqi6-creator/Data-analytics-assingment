```python
import pandas as pd
df = pd.read_csv('Kindle_Book_Dataset.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['title', 'author','save', 'size' , 'publisher']) #dropping cause the too many values were missing or unnecacarry 
```

```python
df.isnull().sum()
```

```python
df['price'] = df['price'].fillna(df['price'].mean())   
 
```

```python
df['pages'] = df['pages'].fillna(df['pages'].median())
```

```python
df['language'] = df['language'].fillna(df['language'].mode()[0])
```

```python
books = df[df['author'] == "James Patterson"]
books.head()
```

```python
df_sorted_price = df.sort_values(by='price', ascending=False)
#Sort the dataset by any column (ascending or descending).
```

```python
df_sorted_price.head()
#Show the top 10 records based on any sorting condition.
```

```python
import matplotlib.pyplot as plt
plt.boxplot(df['price'].dropna())        #BOXPLOT 
plt.ylabel('price')  
plt.show()
```

```python
df['language'].value_counts().plot(kind='bar')
plt.title("Number of Books by Language")
plt.xlabel("Language")                         #BAR CHRT
plt.ylabel("Count")
plt.show()
```
