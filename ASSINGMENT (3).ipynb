```python
import pandas as pd
df = pd.read_csv('Kindle_Book_Dataset.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.shape[0]
```

```python
df = df.dropna(subset=['title', 'author']) #dropping cause many values were missing or null
```

```python
df.isnull().sum()
```

```python
df['publisher'] = df['publisher'].fillna('Unknown')
df['save'] = df['save'].fillna('Unknown')
df['size'] = df['size'].fillna('Unknown')
df.isnull().sum()
```

```python
df.shape[0]
```

```python
df['price'] = df['price'].fillna(df['price'].mean())   
#mean of numerical vaues
```

```python
df['pages'] = df['pages'].fillna(df['pages'].median())
#median of numerical vaues
```

```python
df['language'] = df['language'].fillna(df['language'].mode()[0])
#mode of categoricalvaues
```

```python
books = df[df['author'] == 'James Patterson']
books.head()
 #Filter records by categorical values (e.g., gender = “Female”).
```

```python
df_sorted_price = df.sort_values(by='price', ascending=False)
df_sorted_price.head(10)
#Sort the dataset by any column (ascending or descending). 
```

```python
df[['price','pages']].corr()
#Display the correlation between numerical columns.
```

```python
df['price'].mean() 
#Calculate the average of any numerical column.
```

```python
import matplotlib.pyplot as plt
plt.boxplot(df['price'].dropna())        #BOXPLOT   
plt.title('Box Plot of Book Prices')
plt.xlabel('Book')
plt.ylabel('Price')
plt.show()

```

```python
df['language'].value_counts().plot(kind='bar',color='#a67564')
plt.title("Number of Books by Language")
plt.xlabel('Language')                         #BAR CHRT
plt.ylabel('Count')
plt.show()
```
